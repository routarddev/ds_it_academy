#### New columns: distance (in km), average speed, boolean to know if the flight was late

```python
delays_df['Distance_km'] = delays_df['Distance'] * 1.609344 #km
delays_df['AverageSpeed'] = (60 * delays_df['Distance_km']) / delays_df['ActualElapsedTime'] # km/h

#delay_df['IsArrivalDelay'] = delays_df.apply (lambda row: is_delay(row), axis=1)
delays_df['IsArrivalDelay'] = delays_df.apply(lambda x: True if x['ArrDelay'] > 0 else False, axis=1)
delays_df.head(20)
```

#### Airlines with more cumulative delays

- **UniqueCarrier**: airline
- **sum**: total minutes of delays
- **TotalHoursDelay**: total hours of delays

```python
import numpy as np

airline_delays = delays_df.groupby(['UniqueCarrier'])['ArrDelay'].agg([np.sum])
airline_delays.reset_index(inplace = True)
airline_delays['TotalHoursDelay'] = airline_delays['sum'] / 60
airline_delays.sort_values(['sum'], ascending=False, inplace = True)
airline_delays
```
